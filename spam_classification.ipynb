# IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

```

# IMPORT DATASET

```python
df = pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
df.Category.value_counts()
```

```python
df.isna().sum()
```

```python
df['spam'] = np.where(df['Category'] == 'spam', 1, 0)
```

```python
df.head()
```

```python
df.shape
```

# SPLIT TEST

```python
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
cv = CountVectorizer()

X_train_cv = cv.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.toarray()[:2][0]
```

```python
cv.get_feature_names_out()[1771]
```

```python
cv.vocabulary_
```

```python
X_train_np = X_train_cv.toarray()
X_train_np[0]
```

```python
np.where(X_train_np[0]!=0)
```

```python
X_train_np[0][1771]
```

# MODEL

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = cv.transform(X_test)
```

```python
y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
emails = [
    'Get this for only 20 dollars, this is a special promo',
    'Hi Hans, how are you?'
]

emails_count = cv.transform(emails)
model.predict(emails_count)
```
